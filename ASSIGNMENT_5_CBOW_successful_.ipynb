```python
import numpy as np
import keras.backend as K
from keras.models import Sequential
from keras.layers import Dense, Embedding, Lambda
from keras.utils import np_utils
from keras.preprocessing import sequence
from keras.preprocessing.text import Tokenizer
import gensim
from keras.utils import pad_sequences
import keras as keras
```

```python
#Reading the data
data=open('Covid.txt','r')
cbow_data = [text for text in data if text.count(' ') >= 2]
```

```python
print(cbow_data)
```

```python
#Tokenizing every word using built-in tokenizer
vectorize = Tokenizer()
vectorize.fit_on_texts(cbow_data)
cbow_data = vectorize.texts_to_sequences(cbow_data)
```

```python
print(cbow_data)
```

```python
#Finding total number of words and sentences
total_vocab = sum(len(s) for s in cbow_data)
word_count = len(vectorize.word_index) + 1
print("Total vocabulary :",total_vocab)
print("Word count :",word_count)
```

```python
window_size = 2
```

```python
def cbow_model(data, window_size, total_vocab):
    total_length = window_size*2
    for text in data:
        text_len = len(text)
        for idx, word in enumerate(text):
            context_word = []
            target   = []            
            begin = idx - window_size
            end = idx + window_size + 1
            context_word.append([text[i] for i in range(begin, end) if 0 <= i < text_len and i != idx])
            target.append(word)
            contextual = keras.utils.pad_sequences(context_word, maxlen=total_length)
            final_target = np_utils.to_categorical(target, total_vocab)
            yield(contextual, final_target) 
```

```python
#Building NN model with given parameters
model = Sequential()
model.add(Embedding(input_dim=total_vocab, output_dim=50, input_length=window_size*2))
model.add(Lambda(lambda x: K.mean(x, axis=1), output_shape=(50,)))
model.add(Dense(total_vocab, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
```

```python
from keras.utils import pad_sequences
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
#Training the model
for i in range(10):
  cost = 0
  for contextual, final_target in cbow_model(cbow_data, window_size, total_vocab):
    cost += model.train_on_batch(contextual, final_target)
    print(i, cost)
```

```python
#Creating vector file of some word for testing
dimensions=50
vect_file = open('vectors.txt' ,'w')
vect_file.write('{} {}\n'.format(total_vocab,dimensions))
```

```python
#Accessing the weights of the trained model to write to file
weights = model.get_weights()[0]
for text, i in vectorize.word_index.items():
 final_vec = ' '.join(map(str, list(weights[i, :])))
 vect_file.write('{} {}\n'.format(text, final_vec))
vect_file.close()
```

```python
#Using the vectors created in Gensim
cbow_output = gensim.models.KeyedVectors.load_word2vec_format('vectors.txt', binary=True)
```
